```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
# Reading the dataset and adding the realted headers
filename = "auto.csv"
```

```python
headers= ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels",
          "engine-location","wheel-base","length","width","height","curb-height","engine-type","num-of-cylinders",
          "engine-size","fuel-systems","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg",
          "highway-mpg","price"]
```

```python
# use the Pandas method read_csv() to load the data. Set the parameters "names" equal to the Python list "Headers"
df = pd.read_csv(filename, names= headers)
```

```python
# to see what the data set looks like, we'll use the head() method.
df.head()
```

```python
# Import and handle missing values
# Convert "?" to NaN

import numpy as np
# replace "?" to NaN
df.replace("?", np.nan,inplace=True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
# Caluculate the mean value for the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN" with mean value in "nprmalized-losses"column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Calulate the mean value for the "bore" column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Replace "Nan" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Write your code below and press Shift+Enter to execute
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_bore, inplace=True)
```

```python
# Calculate the mean value for the "horsepower"column
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
df["horsepower"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Calculate the mean value for the "peak-rpm" column
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak-rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#simply drop while row with NaN in "price" column
df.dropna(subset=["price"],axis=0, inplace=True)

#reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
#Convert data types to proper format
df[["bore","stroke"]] = df[["bore","stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

# Data Standardization

```python
df.head()
```

```python
# Convert mpg to L/100km by mathemtical operation (235 divided by mpg)
df['city-L/100km']=235/df["city-mpg"]

#check your transformed data
df.head()
```

```python
# Acc. to the example above, transform mpg to L/100km in the column of "highway-mpg" and change the name to "highway-L/100km"
df["highway-mpg"]=235/df["highway-mpg"]
#rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={"highway-mpg":'highway-L/100km'}, inplace=True)
#check your transferred data
df.head()
```

```python
# replace (original value) by (original value)/(maximum value)
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
```

```python
# Acc to the example above, normalize the column "height".
df['height']=df['height']/df['height'].max()
# show the scaled columns
df[["length","width","height"]].head()
```

```python
# Convert data to correct format
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
# Lets plot the horsepower to see what the distribution of horsepower looks like
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

#set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins =  np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# set group names
group_names = ['Low', 'Medium' , 'High']
```

```python
# aplly the function 'cut' to determine what each value of df['horsepower'] belongs to
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df['horsepower-binned'].value_counts())

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
